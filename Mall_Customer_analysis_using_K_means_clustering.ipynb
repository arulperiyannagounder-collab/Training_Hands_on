```python
q= data.iloc[:,3:].values
q
```

```python
from sklearn.cluster import KMeans
```

```python
cs=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(q)
  cs.append(kmeans.inertia_)
plt.plot(cs)
```

```python
kmns=KMeans(n_clusters=6,init='k-means++',random_state=42)
y_kmeans=kmns.fit_predict(q)
```

```python
kmns.inertia_
```

```python
kmns.labels_
```

```python
plt.scatter(q[y_kmeans==0,0],q[y_kmeans==0,1],color='black',label='c1')
plt.scatter(q[y_kmeans==1,0],q[y_kmeans==1,1],color='red',label='c2')
plt.scatter(q[y_kmeans==2,0],q[y_kmeans==2,1],color='yellow',label='c3')
plt.scatter(q[y_kmeans==3,0],q[y_kmeans==3,1],color='pink',label='c4')
plt.scatter(q[y_kmeans==4,0],q[y_kmeans==4,1],color='blue',label='c5')

plt.scatter(kmns.cluster_centers_[:,0],kmns.cluster_centers_[:,1],color='green',s=300,marker='^')
```
