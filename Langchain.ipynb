```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_classic.chains import RetrievalQA
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
import os
```

```python
# Create dummy file if missing
if not os.path.exists('MCU.txt'):
    with open('MCU.txt', 'w') as f:
        f.write('The Marvel Cinematic Universe (MCU) is a media franchise centering on a series of superhero films.')

loader = TextLoader("MCU.txt")
documents = loader.load()
# Fixed typo: CharracterTextSplitter -> CharacterTextSplitter
splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = splitter.split_documents(documents)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
embeddings=HuggingFaceEmbeddings()
vectorstore=FAISS.from_documents(docs,embeddings)
```

```python
import os
from google.colab import userdata
os.environ['GEMINI_API_KEY'] = userdata.get('MrMercy')
```

```python
llm=ChatGoogleGenerativeAI(model='gemini-pro', temperature=0)
```

```python
qa_chain=RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever()
)
```

```python
question="What is the difference between supervised and unsupervised learning ? "
result = qa_chain.run(question)
print('🤖 Answer :',result)
```
